```python
import yfinance as yf
import pandas as pd

# List of stock tickers
tickers = [
    "GD", "MELI", "BX", "CL", "APD", "PANW", "MTB", "CSX", "CHTR", "SNPS",
    "HBAN", "CDNS", "MPC", "MMM", "FDX", "EPD", "FCX", "SHW", "ICE", "KLAC",
    "USB", "CCI", "MNST", "MCO", "OXY", "ANET", "VMW", "HSY", "PSA", "FTNT"
]

# Define the start and end dates for the data
start_date = "2015-01-01"  # Adjust as needed
end_date = "2023-12-31"    # Adjust as needed

# Create an empty DataFrame to store the closing prices
closing_prices = pd.DataFrame()

# Fetch the weekly closing price for each stock
for ticker in tickers:
    stock_data = yf.download(ticker, start=start_date, end=end_date, interval='1wk')
    closing_prices[ticker] = stock_data['Close']

# Save the DataFrame to a CSV file
closing_prices.to_csv("weekly_closing_prices.csv")
```

The Fama-French Three-Factor Model is described by the following equation:

$$
R_{portfolio} - R_f = \alpha + \beta_{mkt}(Mkt-RF) + \beta_{smb}(SMB) + \beta_{hml}(HML) + \epsilon
$$

Where:

- $R_{portfolio}$ is the return of the portfolio.
- $R_f$ is the risk-free rate.
- $Mkt-RF$ is the market risk premium.
- $SMB$ is the size premium.
- $HML$ is the value premium.
- $\alpha$ is the intercept of the regression.
- $\beta_{mkt}$, $\beta_{smb}$, and $\beta_{hml}$ are the coefficients for the respective factors.
- $\epsilon$ is the error term.


```python
import pandas as pd
import statsmodels.api as sm

# Assuming your DataFrame is named df
# First, calculate the excess returns by subtracting the RF from the portfolio returns
df=merged_df.copy()
df['Excess_Returns'] = df['Returns'] - df['RF']

# Define your independent variables (Mkt-RF, SMB, HML)
X = df[['Mkt-RF', 'SMB', 'HML']]
X = sm.add_constant(X)  # Adds a constant term to the predictor

# Define your dependent variable (Excess_Returns)
y = df['Excess_Returns']

# Create the model
model = sm.OLS(y, X, missing='drop').fit()  # 'missing=drop' handles any NaN values by dropping

# Print out the statistics
model_summary = model.summary()
print(model_summary)
```


1. **Market Risk Premium (Mkt-RF)**: This is the excess return of the market over the risk-free rate. It's the same concept as in CAPM.

2. **Size Premium (SMB for Small Minus Big)**: This factor captures the additional return investors expect from small-cap stocks over large-cap stocks.

3. **Value Premium (HML for High Minus Low)**: This represents the excess return of value stocks (with high book-to-market ratios) over growth stocks (with low book-to-market ratios).

In my regression output:

- The coefficient of 'Mkt-RF' (0.2372) indicates how much the excess returns are expected to move with the market risk premium. A positive coefficient suggests that as the market premium increases, the excess returns are also expected to increase.

- The 'SMB' coefficient (0.1651) shows the return premium associated with small-cap stocks. A positive value suggests a positive relationship between small-cap stocks and the excess returns.

- The 'HML' coefficient (-0.0878) reflects the relationship with the value premium. A negative coefficient here indicates that higher value stocks (as per the model's definition) are associated with lower excess returns in this particular analysis.

The overall R-squared value indicates that these three factors together explain about 6.8% of the variability in the excess returns, which is relatively low. This could mean that there are other factors at play that are not captured by this model. The validity of the model should also be considered in light of the tests for normality of residuals and potential multicollinearity.

```python
import statsmodels.api as sm

master_df['Excess_returns'] = master_df['Returns'] - master_df['RF']
independent_vars = master_df[level_data + first_difference_data + FAMA[['SMB']]]

# Adding a constant (intercept)
independent_vars = sm.add_constant(independent_vars)
model = sm.OLS(master_df['Excess_returns'], independent_vars).fit()
print(model.summary())
```

```python
import statsmodels.api as sm

# Fama-French Model on Excess Returns
fama_factors = ['Mkt-RF', 'SMB', 'HML']
X_fama = sm.add_constant(master_df[fama_factors])
y = master_df['Excess_returns']
model_fama = sm.OLS(y, X_fama).fit()
master_df['Residuals'] = model_fama.resid

# Regression on the Residuals from Fama-French Model
other_vars = level_data + first_difference_data
X_other = sm.add_constant(master_df[other_vars].dropna())
y_residuals = master_df['Residuals'].dropna()
model_residuals = sm.OLS(y_residuals, X_other).fit()

# Print summaries
print("Fama-French Model Summary:")
print(model_fama.summary())
print("\nRegression on Residuals Summary:")
print(model_residuals.summary())
```

The OLS regression results provide insightful information:

1. **Fama-French Model Summary**:

   - None of the Fama-French factors (Mkt-RF, SMB, HML) are statistically significant at typical levels (p > 0.05).
   - The model's overall fit is not statistically significant (F-statistic p-value > 0.05).

2. **Regression on Residuals Summary**:
   - This model explains a higher, but still modest, proportion of the variability in the residuals (R-squared = 0.283).
   - Some variables like 'Bentoil_L', 'VIX_C', 'US_UNEMP_L', and 'US_DEBT_C' are statistically significant.
   - The model's overall fit is statistically significant (F-statistic p-value = 0.00224).
   - However, a large condition number suggests potential multicollinearity issues.

From a financial perspective, the Fama-French factors don't seem to explain much of the variation in excess returns for this dataset. In the second model, specific macroeconomic and market variables provide more explanatory power, indicating their influence over and above the Fama-French factors. The potential multicollinearity issue in the second model, however, suggests that some independent variables might be highly correlated, which could affect the reliability of the coefficients.
